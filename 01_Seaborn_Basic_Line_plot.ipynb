## Seaborn Line Plot
    A Seaborn line plot is used to visualize trends and relationships over a continuous variable, most commonly time. Itâ€™s built on top of Matplotlib but adds statistical intelligence and cleaner styling by default.
    What a Seaborn Line Plot Shows
    - Trend over time or sequence
    - Relationship between two continuous variables
    - Average behavior with variation (confidence interval)

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Tips Dataset 

```python
data = sns.load_dataset("tips")
data_ = pd.DataFrame({"total_bill":data['total_bill'],"tip":data['tip']})
data.head(3)
```

### Example - Relationship between Total Bill and Tip

```python
# Create a line plot showing tip amount vs total bill
sns.lineplot(x="total_bill", y="tip", data=data_)
plt.title("Tip Amount vs Total Bill")
plt.show()
```

### Example - Detailed Sports Watch data display 

```python
# Define sample data for a sports watch pulse analysis
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
data_pulse = pd.DataFrame({"Average Pulse": x, "Calorie Burnage": y})

# Plot pulse vs calories with custom markers and titles
sns.lineplot(x="Average Pulse", y="Calorie Burnage", data=data_pulse, marker="o")

# Customize appearance
plt.title("Sports Watch Data", color="darkred", size=18)
plt.xlabel("Average Pulse (BPM)")
plt.ylabel("Calorie Burnage (kcal)")
plt.show()
```

### Example - Annual Sales Trend Comparison [ Company A vs Company B ]

```python
# Create DataFrame
data = pd.DataFrame({                        
    "Year": [2016, 2017, 2018, 2019, 2020],   # Year values (x-axis)
    "Sales-A": [100, 120, 150, 170, 200],    # Sales data for Company A
    "Sales-B": [80, 110, 130, 160, 180]      # Sales data for Company B
})

sns.lineplot(x="Year",y="Sales-A",data=data,label="Company A", marker="o")  # Plot line for Company A
sns.lineplot( x="Year",y="Sales-B",data=data,label="Company B", marker="o")  # Plot line for Company B
font1 = {'family': 'serif', 'color': 'darkred', 'size': 18}   # Title font style
font2 = {'family': 'serif', 'color': 'darkred', 'size': 15}   # Axis label font style
plt.title("Annual Sales Trend Comparison [Company A vs Company B]", fontdict=font1)  # Plot title
plt.xlabel("Year", fontdict=font2)                     # X-axis label
plt.ylabel("Sales", fontdict=font2)                    # Y-axis label
plt.legend(loc=2)                                      # Legend position (upper left)
plt.show()                                             
```
