## Seaborn PAIR PLOT
    A pair plot is a grid of plots that shows pairwise relationships between numerical features in a dataset.
    - Off-diagonal plots → scatter plots [ feature vs feature ]
    - Diagonal plots → distribution of each feature [ histogram or KDE ]
    - Helps quickly understand:
      - Relationships between variables
      - Class separation [ when hue is used ]
      - Correlation patterns
      - Outliers and data spread
    - It is mainly used during EDA [ Exploratory Data Analysis ].

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

```python
iris = sns.load_dataset("iris")
iris.head(2)
```

### Example - Basic Pair Plot [ Default ]

```python
# Basic pair plot
sns.pairplot(iris)
plt.show()
```

### Example - Pair Plot with Hue [ Best for Understanding Classes ]

```python
sns.pairplot(iris,hue = "species")
plt.show()
```

### Example - Pair Plot with KDE on Diagonal

```python
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.show()
```

### Example - Pair Plot with Custom Scatter Style

```python
sns.pairplot(
    iris,
    hue="species",
    markers=["o", "s", "D"],  # different marker styles
    plot_kws={"alpha": 0.7}   # transparency for overlap
)
plt.show()
```
