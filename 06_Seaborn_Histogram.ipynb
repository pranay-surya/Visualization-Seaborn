## Seaborn Histogram
    A Seaborn histogram is used to visualize the distribution of a numerical variable by grouping data into bins and showing how frequently values occur in each bin. It helps understand the shape, spread, skewness, and outliers in data.
    What a Seaborn Histogram Shows ::
    - X-axis → Numerical variable
    - Y-axis → Frequency or count
    - Bars (bins) → Range of values
    - Optional KDE curve → Smooth distribution estimate

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Loading Tips Dataset 

```python
data= sns.load_dataset("tips")
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':13}
data.head(2)
```

### Example - Distribution of Total Bill

```python
# Histogram showing distribution of total bill amounts
# bins = Number of intervals
sns.histplot(data=data, x="total_bill", bins=20,color = 'green')

# Title and axis labels of histogram plot 
plt.title("Distribution of Total Bill Amount",fontdict=font1)
plt.xlabel("Total Bill Amount",fontdict=font2)
plt.ylabel("Frequency",fontdict=font2)
plt.show()
```

### Example - Distribution of Tip Amount

```python
# Histogram showing how tip amounts are distributed
sns.histplot(data=data,x="tip",bins=20,color = 'green')

# Title and axis labels of histogram plot
plt.title("Distribution of Tip Amount",fontdict=font1)
plt.xlabel("Tip Amount",fontdict=font2)
plt.ylabel("Frequency",fontdict=font2)
plt.show()
```

### Example - Total Bill Distribution with KDE Curve

```python
# Histogram with KDE curve to show smooth distribution
# kde =  Add density curve
sns.histplot(data=data,x="total_bill",kde=True,color = 'green')

# Title and axis labels of histogram plot 
plt.title("Total Bill Distribution with KDE",fontdict=font1)
plt.xlabel("Total Bill Amount",fontdict=font2)
plt.ylabel("Density",fontdict=font2)
plt.show()
```

### Example - Tip Distribution by Meal Time

```python
# Histogram comparing tip distribution for lunch and dinner
# Lunch vs Dinner
# ' multiple="dodge" ' = Side-by-side bars
sns.histplot(data=data,x="tip",hue="time",multiple="dodge")

# Title and axis labels
plt.title("Tip Distribution by Meal Time",fontdict=font1)
plt.xlabel("Tip Amount",fontdict=font2)
plt.ylabel("Frequency",fontdict=font2)
plt.show()
```

### Example - Total Bill Distribution by Gender

```python
# Histogram comparing total bill distribution by gender
# hue="sex" ->  Gender-based comparison
# multiple="stack" -> Stack bars for comparison
sns.histplot(data=data,x="total_bill",hue="sex",multiple="stack')

# Title and axis labels of histogram plot
plt.title("Total Bill Distribution by Gender",fontdict=font1)
plt.xlabel("Total Bill Amount",fontdict=font2)
plt.ylabel("Count",fontdict=font2)
plt.show()
```
