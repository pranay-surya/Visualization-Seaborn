## Seaborn Violin Plot
    A Seaborn violin plot is a statistical visualization tool used to compare the distribution of data across different categories. It is essentially a combination of a Box Plot and a Kernel Density Estimate (KDE).
    Violin plot shows ::
    - Distribution shape of the data (density)
    - Median and quartiles (like a box plot)
    - Spread and skewness of the data

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Loading Tips dataset 

```python
tips = sns.load_dataset("tips")
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':13}
tips.head(2)
```

### Example - Tip Distribution by Gender

```python
# Shows how tip amounts are distributed for Male vs Female
# Width indicates where values are more concentrated
sns.violinplot(x="sex", y="tip", data=tips)
plt.title("Distribution of Tips by Gender",fontdict=font1)
plt.xlabel("Sex",fontdict=font2)
plt.ylabel("Tip",fontdict=font2)
plt.show()
```

### Example - Total Bill Distribution by Day

```python
# Helps compare spending patterns across days
# Wider sections indicate frequent bill values
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill Distribution Across Days",fontdict=font1)
plt.xlabel("Day",fontdict=font2)
plt.ylabel("Total_Bill",fontdict=font2)
plt.show()
```

### Example - Tip Distribution by Time [ Lunch vs Dinner ]

```python
sns.violinplot(x="time", y="tip", data=tips)
plt.title("Tip Distribution by Time",fontdict=font1)
plt.xlabel("Time",fontdict=font2)
plt.ylabel("Tip",fontdict=font2)
plt.show()
```

### Example - Tip Distribution by Day and Gender [ Split Violin ]

```python
# Split violins allow side-by-side comparison
# Easy to compare male vs female tipping behavior on each day
# Very useful for multi-category analysis
sns.violinplot(x="day", y="tip", hue="sex", data=tips, split=True)
plt.title("Tip Distribution by Day and Gender",fontdict=font1)
plt.xlabel("Day",fontdict=font2)
plt.ylabel("Tip",fontdict=font2)
plt.show()
```
