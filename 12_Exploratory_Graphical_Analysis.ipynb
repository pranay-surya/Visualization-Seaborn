```python
font1 = {
    'family': 'serif',
    'color': 'darkred',
    'weight': 'bold',
    'size': 16
}

font2 = {
    'family': 'serif',
    'color': 'darkred',
    'size': 12
}
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_theme()


# Load dataset
tips = sns.load_dataset("tips")

# Create subplot grid (2 columns)
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(16, 26))
axes = axes.flatten()

# 1. Multi-line Plot
sns.lineplot(x="day", y="total_bill", hue="sex", data=tips, ax=axes[0])
axes[0].set_title("Multi-line Plot: Total Bill by Day & Gender", fontdict=font1)
axes[0].set_xlabel("Day", fontdict=font2)
axes[0].set_ylabel("Total Bill", fontdict=font2)

# 2. Scatter Plot
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips, ax=axes[1])
axes[1].set_title("Scatter Plot: Total Bill vs Tip", fontdict=font1)
axes[1].set_xlabel("Total Bill", fontdict=font2)
axes[1].set_ylabel("Tip", fontdict=font2)

# 3. Histogram
sns.histplot(tips["total_bill"], bins=20, ax=axes[2],color='green',kde=True)
axes[2].set_title("Histogram: Total Bill Distribution", fontdict=font1)
axes[2].set_xlabel("Total Bill", fontdict=font2)
axes[2].set_ylabel("Frequency", fontdict=font2)

# 4. Bar Plot
sns.barplot(x="day", y="tip", data=tips, ax=axes[3],color='green',width=0.4)
axes[3].set_title("Bar Plot: Average Tip by Day", fontdict=font1)
axes[3].set_xlabel("Day", fontdict=font2)
axes[3].set_ylabel("Average Tip", fontdict=font2)

# 5. Box Plot
sns.boxplot(x="day", y="total_bill", data=tips, ax=axes[4],color='green')
axes[4].set_title("Box Plot: Total Bill by Day", fontdict=font1)
axes[4].set_xlabel("Day", fontdict=font2)
axes[4].set_ylabel("Total Bill", fontdict=font2)

# 6. Violin Plot
sns.violinplot(x="day", y="tip", data=tips, ax=axes[5],color='green')
axes[5].set_title("Violin Plot: Tip Distribution by Day", fontdict=font1)
axes[5].set_xlabel("Day", fontdict=font2)
axes[5].set_ylabel("Tip", fontdict=font2)

# 7. KDE Plot
sns.kdeplot(data=tips, x="total_bill", fill=True, ax=axes[6],color='green')
axes[6].set_title("KDE Plot: Total Bill Density", fontdict=font1)
axes[6].set_xlabel("Total Bill", fontdict=font2)
axes[6].set_ylabel("Density", fontdict=font2)

# 8. Heatmap
corr = tips.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[7])
axes[7].set_title("Heatmap: Correlation Matrix", fontdict=font1)

# 9. Regression (LM) Plot
sns.regplot(x="total_bill", y="tip", data=tips, ax=axes[8],color='green')
axes[8].set_title("LM Plot: Total Bill vs Tip", fontdict=font1)
axes[8].set_xlabel("Total Bill", fontdict=font2)
axes[8].set_ylabel("Tip", fontdict=font2)

# 10 Count plot
sns.countplot(x="smoker", hue="sex", data=tips, ax=axes[9],width=0.3)
axes[9].set_title("Count Plot: Smoker Distribution by Gender", fontdict=font1)
axes[9].set_xlabel("Smoker", fontdict=font2)
axes[9].set_ylabel("Count", fontdict=font2)

# Layout adjustment
plt.tight_layout()
plt.show()
#plt.savefig("Seaborn_Final")
```
