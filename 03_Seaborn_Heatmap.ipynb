## Matplotlib Heatmap
    A heatmap is a data visualization technique that uses colors to represent values in a matrix or table. It helps to quickly identify patterns, trends, correlations, or outliers in data.
    Heatmaps are useful for ::
    - Correlation analysis (feature relationships)
    - Confusion matrices (classification performance)
    - Pattern detection (clusters, anomalies)
    - Comparing values across two dimensions


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.style as style 
import random
style.use("ggplot")
```

### Example - Basic Heatmap

```python
data = np.random.rand(6,6)
sns.heatmap(data,annot = True,cmap="YlGnBu") # annot = 'True' shows the actual values inside boxesa
plt.show()
```

### Example - Correlation Heatmap 
     A correlation heatmap shows how strongly features are related to each other.
     Values range between -1 and +1:
     +1 → Perfect positive relation (if one increases, other increases).
     -1 → Perfect negative relation (if one increases, other decreases).
      0 → No relation.
     In the heatmap →
     Dark red/blue = strong correlation
     Light = weak or no correlation

```python
## Sample dataset
data = pd.DataFrame({
    "Math": [85, 90, 78, 92, 88],
    "Science": [80, 85, 75, 95, 90],
    "English": [70, 65, 80, 75, 72]
})
corr = data.corr()
# df.corr() = it's used to calculate the correlation matrix between all numeric columns in a DataFrame.
# Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

### Example - Missing Values Heatmap
      Real datasets often have NaN (missing values).
      Heatmaps are very useful to visualize missing data.
     Example: seaborn.heatmap(df.isnull())
     White boxes = missing values
     Dark boxes = filled values

```python
## Example dataset with missing values
data = {
    "A": [1, 2, np.nan, 4,5,8,np.nan],
    "B": [5, np.nan, 7, 8,1,3,np.nan],
    "C": [9, 10, 11, np.nan,np.nan,np.nan,23]
}
df = pd.DataFrame(data)
sns.heatmap(df.isnull(), cmap="YlGnBu", cbar=False)
plt.show()
```

### Example - Cluster Heatmap
        A cluster heatmap is like a normal heatmap, but with clustering applied.
        It not only shows values as colors,
        But also groups similar rows and columns together using hierarchical clustering.
        
        Dendrogram (Branch structure / Tree) – The “branch type thing” you saw.
        It shows how similar rows (or columns) are.
        Short branches = very similar, long branches = less similar.

```python
## Random dataset 
np.random.seed(42)
data = pd.DataFrame({
    "Math": np.random.randint(40, 100, 10),
    "Physics": np.random.randint(35, 95, 10),
    "Chemistry": np.random.randint(45, 98, 10),
    "English": np.random.randint(30, 90, 10),
    "History": np.random.randint(25, 85, 10)})
 
corr = data.corr()
sns.clustermap(corr, cmap="coolwarm", annot=True,figsize=(5,5))
plt.show()
```
