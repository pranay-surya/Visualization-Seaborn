## Seaborn Scatter Plot 
    A Seaborn scatter plot is used to visualize the relationship between two numerical variables. It’s great for spotting patterns, trends, clusters, and outliers—and it looks cleaner than plain Matplotlib with less code.
    To check correlation between features ::
    - To visualize clusters
    - To detect outliers
    - Before applying regression or ML models

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Tips Dataset

```python
tips= sns.load_dataset("tips")
tips.head(2)
```

```python
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':15}
```

### Example - Basic Scatter plot [ Total Bill vs Tip ]

```python
# Scatter plot showing the relationship between total bill and tip amount
sns.scatterplot(x = "total_bill",y = "tip",data= tips)
 
# Title and label of the plot
plt.title("Total Bill vs Tip ",fontdict = font1)
plt.xlabel("Total_bill",fontdict = font2)
plt.ylabel("Tips",fontdict = font2)
plt.show()
```

### Example - Tips vs Total Bill [ Gender-wise ]

```python
# Scatter plot showing the relationship between total bill and tip amount
# Points are colored based on customer gender
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex")

# Title and label of the plot
plt.title("Total Bill vs Tip by Gender", fontdict=font1)
plt.xlabel("Total Bill Amount", fontdict=font2)
plt.ylabel("Tip Amount", fontdict=font2)
plt.show()
```

### Example - Total Bill vs Tip with Group Size Representation

```python
# Scatter plot showing relationship between total bill and tip
# Each point represents a single customer visit
sns.scatterplot(x="total_bill", y="tip",data=tips,size="size")

# Title and label of the scatter plot
plt.title("Total Bill vs Tip with Group Size Representation", fontdict=font1)
plt.xlabel("Total Bill Amount", fontdict=font2)
plt.ylabel("Tip Amount", fontdict=font2)
plt.show()
```

### Example - Total Bill vs Tip Analysis by Day, Gender & Party Size

```python
# Scatter plot showing the relationship between total bill amount and tip
# Each point represents a customer visit
sns.scatterplot( x="total_bill",y="tip",data=tips,hue="day",style="sex", size="size" )

# Title and label of the plot
plt.title("Total Bill vs Tip Analysis by Day, Gender & Party Size", fontdict=font1)
plt.xlabel("Total Bill Amount", fontdict=font2)
plt.ylabel("Tip Amount", fontdict=font2)
plt.show()
```
