## Seaborn Box Plot
    A Seaborn box plot (also called a box-and-whisker plot) is a statistical visualization used to summarize the distribution of a numerical variable and compare distributions across categories.
    A single box plot captures five key statistics ::
    - Minimum :- smallest value (excluding outliers)
    - Q1 (25th percentile) :– lower edge of the box
    - Median (50th percentile) :– line inside the box
    - Q3 (75th percentile) :– upper edge of the box
    - Maximum :– largest value (excluding outliers)

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Loading Tips Dataset 

```python
tips = sns.load_dataset("tips")
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':13}
tips.head(3)
```

### Example - To visualize outliers in total_bill

```python
# Create a box plot to visualize outliers in total_bill
sns.boxplot(x="day", y="total_bill", data=tips)

# Add a title and label for box plot 
plt.title("Outlier Detection in Total Bill by Day",fontdict=font1)
plt.xlabel("Days",fontdict=font2)
plt.ylabel("Total_Bills",fontdict=font2)
plt.show()
```

### Example - Tip Amount by Gender

```python
# Create a box plot to compare tip amounts by gender
sns.boxplot(x="sex", y="tip", data=tips)

# Add title and label for box plot 
plt.title("Tip Distribution by Gender",fontdict=font1)
plt.xlabel("Sex",fontdict=font2)
plt.ylabel("Tips",fontdict=font2)
plt.show()
```

### Example - Total Bill by Smoker Status

```python
# Box plot to analyze spending differences between smokers and non-smokers
sns.boxplot(x="smoker", y="total_bill", data=tips)

# Add plot title and label 
plt.title("Total Bill Distribution by Smoker Status",fontdict=font1)
plt.xlabel("Smoker",fontdict=font2)
plt.ylabel("Total_Bill",fontdict=font2)
plt.show()
```

### Example - Tip Distribution by Time [ Lunch vs Dinner ]

```python
# Create box plot for tip amount across meal times
sns.boxplot(x="time", y="tip", data=tips)

# Add title and label for box plot 
plt.title("Tip Distribution by Meal Time",fontdict=font1)
plt.xlabel("Time",fontdict=font2)
plt.ylabel("Tips",fontdict=font2)
plt.show()
```

### Example - Total Bill by Day with Gender Comparison [ Hue ] 

```python
# Box plot with an additional categorical comparison using hue
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)

# Add title and label to explain multi-category comparison
plt.title("Total Bill by Day and Gender",fontdict=font1)
plt.xlabel("Day",fontdict=font2)
plt.ylabel("Total_Bill",fontdict=font2)
plt.show()
```
