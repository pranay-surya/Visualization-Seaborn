## Bar plot
    A Seaborn bar plot is used to visualize the relationship between a categorical variable and a numerical variable.It displays aggregate values [ by default, the mean ] for each category using rectangular bars.
    What a Seaborn Bar Plot Shows ::
    - Each bar represents a category
    - The height of the bar represents a statistical summary [ mean by default ]
    - Error bars show confidence intervals [ usually 95% ]

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Loading Tips Dataset

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':15}
```

### Example - Average Total Bill by Day

```python
# Bar plot showing the average total bill for each day
# By default, Seaborn calculates the mean of 'total_bill' for each day
sns.barplot(x="day",y="total_bill",data=tips,width=0.4)

# Title and label of the bar plot
plt.title("Average Total Bill by Day",fontdict=font1)
plt.ylabel("Total_bills",fontdict=font2)
plt.xlabel("Days",font2)
plt.show()
```

### Example - Average Total Bill by Day and Gender

```python
# Grouped bar plot showing the average total bill for each day
# Bars are divided based on customer gender
sns.barplot(x="day",y="total_bill",data=tips,hue="sex")

# Title and label of the bar plot
plt.title("Average Total Bill by Day and Gender",fontdict=font1)
plt.ylabel("Total_bills",fontdict=font2)
plt.xlabel("Days",font2)
plt.show()
```

### Example - Sum of Total Bills Across Days

```python
# Horizontal bar plot showing the total (sum) of bills for each day
# Unlike the default mean, estimator=sum aggregates total_bill values
sns.barplot( y="day",x="total_bill",data=tips,estimator=sum,width=0.4)

# Title and label of the bar plot
plt.title("Total Bill Amount by Day",fontdict=font1)
plt.xlabel("Total_bills",fontdict=font2)
plt.ylabel("Days",font2)
plt.show()
```

```python
# Create a figure with multiple subplots
plt.figure(figsize=(18, 12))
plt.suptitle("Exploratory Analysis of Customer Spending and Tips",family='serif',color='darkred',size=18)


# 1. Average Total Bill by Day
plt.subplot(2, 3, 1)
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Total Bill by Day")

# 2. Average Tip by Day
plt.subplot(2, 3, 2)
sns.barplot(x="day", y="tip", data=tips)
plt.title("Average Tip by Day")

# 3. Average Total Bill by Gender
plt.subplot(2, 3, 3)
sns.barplot(x="sex", y="total_bill", data=tips)
plt.title("Average Total Bill by Gender")

# 4. Average Tip by Day and Gender
plt.subplot(2, 3, 4)
sns.barplot(x="day", y="tip", hue="sex", data=tips)
plt.title("Average Tip by Day and Gender")
plt.legend(title="Gender")

# 5. Total Revenue by Day (sum)
plt.subplot(2, 3, 5)
sns.barplot(x="day", y="total_bill", data=tips, estimator=sum)
plt.title("Total Revenue by Day")

# 6. Average Total Bill by Party Size
plt.subplot(2, 3, 6)
sns.barplot(x="size", y="total_bill", data=tips)
plt.title("Average Total Bill by Party Size")
plt.tight_layout()
plt.show()
```
