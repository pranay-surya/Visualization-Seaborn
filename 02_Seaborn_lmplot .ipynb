## Seaborn lmplot
    lmplot in Seaborn is used to visualize the relationship between two numerical variables by combining a scatter plot with a regression line. It helps you understand trends, patterns, and correlations in data.
    When to Use lmplot ::
    - To study relationships between two numeric variables
    - To compare trends across categories
    - To check linear vs non-linear patterns
    - For quick exploratory data analysis (EDA)

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Loading Titanic Dataset 

```python
titanic = sns.load_dataset("titanic")
titanic.head(2)
```

```python
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':15}
```

### Example - Survival Analysis [ Double Line Plot ]

```python
# Survival Rate vs. Age [ Male vs Female ]
# => 1 = Alive 
# => 0 = Dead
# Seaborn smoth line plot 
sns.lineplot(x = "age",y = "survived", hue = "sex",data = titanic)
plt.title("Survival Rate VS Age [ Male vs Female ] ",fontdict= font1)
plt.xlabel("Age",fontdict = font2)
plt.ylabel("Survival Rate",fontdict = font2)
plt.show()
```

```python
# Seaborn lmplot Double Lineplot 
# 1 = Alive and 0 = Dead
# lowess=True creates a non-parametric smooth line without assuming a linear relationship
sns.lmplot(x = "age",y = "survived", hue = "sex",data = titanic,lowess=True,height = 3,aspect = 1.5)
plt.title("Survival Rate VS Age [ Male vs Female ] ",fontdict= font1)
plt.xlabel("Age",fontdict = font2)
plt.ylabel("Survival Rate",fontdict = font2)
plt.show()
```

```python
# Smooth Sirvival curve by passenger class
# 1 = Alive and 0 = Dead
# lowess=True creates a non-parametric smooth line without assuming a linear relationship
sns.lmplot(x = "age",y = "survived", hue = "pclass",data = titanic,lowess=True, height = 3,aspect = 1.5)
plt.title("Survival Rate VS Age [ Male vs Female ] ",fontdict= font1)
plt.xlabel("Age",fontdict = font2)
plt.ylabel("Survival Rate",fontdict = font2)
plt.show()
```

```python
# Smooth Multiple - Condition Lineplot 
# Survival According to passenger class [ Male vs Female ]
# lowess=True creates a non-parametric smooth line without assuming a linear relationship
sns.lmplot(x = "age",y = "survived", hue = "sex",col="pclass",data = titanic,lowess=True,height = 5,aspect = 1.5)
plt.subplots_adjust(top = 0.85)
plt.show()
```

```python
# Combine Smooth Lineplot
# Create a unique grouping column by combining sex and pclass
# This allows the hue mapping to distinguish groups like 'male_class1', 'female_class3', etc.
# lowess = True creates a non-parametric smooth line without assuming a linear relationship
titanic['group'] = titanic["sex"]+"_class"+titanic["pclass"].astype(str)
sns.lmplot(x = "age",y = "survived", hue = "group",data = titanic,lowess=True, height = 3,aspect = 1.5)
plt.title("Survival Rate VS Age [ Male vs Female ] ",fontdict= font1)
plt.xlabel("Age",fontdict = font2)
plt.ylabel("Survival Rate",fontdict = font2)
plt.show()
```
