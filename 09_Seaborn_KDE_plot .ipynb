## KDE Plot [ Kernel Density Estimation ]
    A Seaborn KDE plot visualizes the probability density of a continuous variable by smoothing the data into a continuous curve. Instead of counting bars like a histogram, it estimates how data is distributed across values using Kernel Density Estimation.
    What it shows ::
    - The shape of the data distribution (peaks, spread, skewness)
    - Density on the y-axis (not counts)
    - Smooth trends without binning artifacts

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Loading tips Dataset 

```python
tips = sns.load_dataset("tips")
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':13}
tips.head(2)
```

### Example - Basic KDE Plot & KDE plot with adjusted bandwidth [ Single Variable ]

```python
plt.figure(figsize=(14,5))
# KDE plot for total bill amount
plt.subplot(1,2,1)
sns.kdeplot(data=tips, x="total_bill",fill=True,color='green')

# Title and lable for KDE plot 
plt.title("KDE Plot of Total Bill",fontdict=font1)
plt.xlabel("Total Bill",fontdict=font2)
plt.ylabel("Density",fontdict=font2)

plt.subplot(1,2,2)
# KDE plot with adjusted bandwidth
sns.kdeplot(data=tips, x="total_bill", bw_adjust=0.5,color='green',fill=True)

# Title and lable for KDE plot 
plt.title("KDE Plot of Total Bill",fontdict=font1)
plt.xlabel("Total Bill",fontdict=font2)
plt.ylabel("Density",fontdict=font2)
plt.show()
```

### Example - KDE Plot with Hue [ Group Comparison ]

```python
# KDE plot comparing males and females
sns.kdeplot(data=tips, x="total_bill", hue="sex", fill=True)

# Title and lable for KDE plot 
plt.title("KDE Plot of Total Bill",fontdict=font1)
plt.xlabel("Total Bill",fontdict=font2)
plt.ylabel("Density",fontdict=font2)
plt.show()
```

### Example - Basic KDE Contour [ Clean Circular Lines ]

```python
sns.kdeplot( data=tips, x="total_bill", y="tip", levels=6,color="green",fill=True)

# Title and lable for KDE plot 
plt.title("KDE Contour Plot: Total Bill vs Tip",fontdict=font1)
plt.xlabel("Total Bill",fontdict=font2)
plt.ylabel("Tip",fontdict=font2)
plt.show()
```

### Example - KDE + Scatter Overlay 
### Example - KDE Contour by Gender [ Group-wise Circles ]
### Example - KDE Contour with Threshold [ Noise Removal ]

```python
plt.figure(figsize=(14,5))

## KDE + Scatter Overlay 
plt.subplot(1,3,1)
sns.scatterplot( data=tips, x="total_bill", y="tip",alpha=0.4)
sns.kdeplot(data=tips,x="total_bill",y="tip",levels=6,color="green")
plt.title("Scatter + KDE Contours :: Total Bill vs Tip",fontdict=font2)
plt.xlabel("Total Bill",fontdict=font2)
plt.ylabel("Tip",fontdict=font2)

## KDE Contour by Gender [ Group-wise Circles ]
plt.subplot(1,3,2)
sns.kdeplot( data=tips, x="total_bill", y="tip", hue="sex", levels=6)
plt.title("KDE Contours by Gender",fontdict=font2)
plt.xlabel("Total Bill",fontdict=font2)
plt.ylabel("Tip",fontdict=font2)

## KDE Contour with Threshold [ Noise Removal ]
plt.subplot(1,3,3)
sns.kdeplot( data=tips, x="total_bill", y="tip", fill=True, levels=10, thresh=0.1)
plt.title("KDE with Density Threshold",fontdict=font2)
plt.xlabel("Total Bill",fontdict=font2)
plt.ylabel("Tip",fontdict=font2)
plt.show()
```
