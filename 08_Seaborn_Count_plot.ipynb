## Seaborn Count Plot
    A Seaborn count plot is a simple but powerful categorical plot used to show the frequency (count) of observations in each category of a variable.
    Use a count plot when ::
    - Your variable is categorical
    - You want to understand distribution of categories
    - You want a quick overview of imbalance in data

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.style as style 
import seaborn as sns
style.use("ggplot")
```

### Loading Tips Dataset

```python
tips = sns.load_dataset("tips")
font1 = {'family':'serif','color':'darkred','size':18}
font2 = {'family':'serif','color':'darkred','size':13}
tips.head(2)
```

### Example - Count of customers by day

```python
# Count plot for number of customers on each day
sns.countplot(x="day", data=tips, color = 'green',width=0.5)

# Add title and labels
plt.title("Customer Count by Day",fontdict=font1)
plt.xlabel("Day",fontdict=font2)
plt.ylabel("Number of Customers",fontdict=font2)
plt.show()
```

### Example - Gender wise customer count

```python
# Count plot for gender distribution
sns.countplot(x="sex", data=tips,color='green',width=0.4)

# Add title and labels
plt.title("Gender Distribution of Customers",fontdict=font1)
plt.xlabel("Gender",fontdict=font2)
plt.ylabel("Count",fontdict=font2)
plt.show()
```

### Example - Day wise count split by gender [ using hue ]

```python
# Count plot with hue for comparison
sns.countplot(x="day", hue="sex", data=tips, palette=["orange", "green"])

# Add title and labels
plt.title("Customer Count by Day and Gender",fontdict=font1)
plt.xlabel("Day",fontdict=font2)
plt.ylabel("Count",fontdict=font2)
plt.show()
```

### Example - Smoker vs Non-Smoker count by time

```python
# Count plot with hue and custom orientation
sns.countplot(x="time", hue="smoker", data=tips,width=0.4,palette = ["green", "orange"])

# Add title and labels
plt.title("Smoker vs Non-Smoker Count by Time",fontdict=font1)
plt.xlabel("Time [ Lunch / Dinner ]",fontdict=font2)
plt.ylabel("Number of Customers",fontdict=font2)
plt.show()
```
